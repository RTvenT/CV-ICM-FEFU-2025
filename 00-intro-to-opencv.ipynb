# __Введение в OpenCV__

## __Импорт и просмотр изображений__

```python
import cv2
import numpy as np
```

```python
ORIG_PATH = './images/orig/'
NEW_PATH = './images/new/'
```

Для загрузки изображения используется ф-ия `cv2.imread()`

```python
image = cv2.imread(ORIG_PATH + "cat.jpeg")
```

```python
cv2.imshow("Window Name", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## __Сохранение изображений__

`cv2.imwrite(filename, image, params=None)` - основной инструмент для сохранения изображений в OpenCV. Возвращает булевое значение, об успешности сохранения.

Параметры:
- `filename` - строка с именем файла и расширением
- `image` - массив NumPy (изображение для сохранения)
- `params` - дополнительные параметры кодировки (опционально)

```python
cv2.imwrite(NEW_PATH + "cat_writen.jpeg", img=image)
```

## __Кадрирование и масштабирование__

Кадрирование - это выделение и сохранение определенной области изображения. В OpenCV это делается с помощью срезов NumPy массива.

`cropped = image[y_start:y_end, x_start:x_end]`

```python
image = cv2.imread(ORIG_PATH + "bloon.jpg")
```

```python
x, y = 100, 50
w, h = 300, 400

cropped = image[y:y+h, x:x+w]

cv2.imshow("Original", image)
cv2.imshow("Cropped", cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_b, img_g, img_r = cv2.split(image)

cv2.imwrite(NEW_PATH + "bloon_b.jpg", img_b)
cv2.imwrite(NEW_PATH + "bloon_g.jpg", img_g)
cv2.imwrite(NEW_PATH + "bloon_r.jpg", img_r)
```

```python
image_bloon_croped = image[145:430, 290:540]
resized = cv2.resize(image_bloon_croped, image.shape[:-1])

cv2.imwrite(NEW_PATH + "croped_resized_bloon.jpg", resized)

cv2.imshow("Original", image)
cv2.imshow("Cropped", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
